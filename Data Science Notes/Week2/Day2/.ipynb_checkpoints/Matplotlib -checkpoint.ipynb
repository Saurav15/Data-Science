# Plotting a Bar-Graph

### Why we use Bar-Graphs ?
#### Used to compare things between groups .
#### Also when we are trying to measure changes over time Bar-Graphs are very well suited


```python
import matplotlib.pyplot as plt
```

```python
plt.bar([1,3,5,7,9],[5,2,7,8,2],label='Example-One')
```

```python
# we can also plot multiple data on the same Graphs
plt.bar([1,3,5,7,9],[5,2,7,8,2],label='Example-One')
plt.bar([2,4,6,8,10],[5,2,7,8,2],label='Example-One')
plt.title('Info')
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True,color='k')
```

# Plotting Histogram 
## What is the difference between the Bar-Graph and the Histogram ?
In Histograms we ahve quantitative variables and in Bar-Graph they have categorical variables .
Eg. If i ant to plot GDP grwth of every city in the Country , at that time i will use the bar plot because it has a category 
     in the same example if i wanna plot a quantitayive data like the age group at that time i will be using Histogram.

```python
import matplotlib.pyplot as plt 
```

```python
population_ages = [22,55,62,76,34,54,87,94,12,59,78,123,110,100,94]
bins = 12 # The bins parameter tells you the number of bins that your data will be divided into. You can specify it as an integer or as a list of bin edges.

#Bins are the number of intervals you want to divide all of your data into, such that it can be displayed as bars on a histogram. A simple method to work our how many bins are suitable is to take the square root of the total number of values in your distribution.

# can also be bins = [0,10,20,30,40,50,60,80,100,130] Try this 
```

```python
plt.hist(population_ages,bins , histtype = 'bar', rwidth=0.8)
plt.title("Histogram")
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.legend()
```

# Plotting Scatter Graph :

## Why do we need scatter plot ?
This plotting of points is called scatter plot because we are scattering all the points from our dataset on a map. Scatter plots are used to understand the relationships between variables.
For example, we try to understand whether there is any relationship between the increase or decrease in the value of one variable.

```python
import matplotlib.pyplot as plt
```

```python
# Lets define some varibles to use
x = [1,2,3,4,5,6]
y = [5,3,7,4,8,1]
```

```python
plt.scatter(x,y,label='Scatter-Graph' , color='k')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title("Graph")
plt.grid(True,color='k')
```

# Plotting a Stack Graph
They are very similer to line graphs.
can be used to track changes over time for one or more groups
They are good to use when u are tracking changes between two or more groups , that make up one whole category .

```python
import matplotlib.pyplot as plt
```

```python
# These are some variables that we need to compare & they can be compares as they belong to the same category !
sleeping = [2,7,3,8,3,1]
eating =   [7,3,8,3,5,2]
working =  [1,4,2,5,2,6]
playing =  [8,3,1,5,6,2]

days=[1,2,3,4,5,6]
```

```python
# the problem here is we cannot have label for our data !
# so what we can do is we can just nake the empty linear graphs and label them !
plt.plot([],[],label='sleeping',color='m',linewidth=4)
plt.plot([],[],label='eating',color='c',linewidth=4)
plt.plot([],[],label='working',color='r',linewidth=4)
plt.plot([],[],label='playing',color='k',linewidth=4)


# Now lrts plot the Stack
plt.stackplot(days,sleeping,working,eating,playing,colors=['m','c','r','k'])
plt.legend()
```

# Plotting Pie-Chart
Pie chart is similar to the Stack Plot . They are many times used to show the percentage share .

```python
import matplotlib.pyplot as plt
```

```python
# Now here we only have one data i.e we dont need 2 data variables like x,y 
# So lets give some data first
slices = [2,8,4,5]
activities = ['Eating','Sleeping','Studying','Playing']
```

```python
# Now we have the data .... So lets plot
plt.pie(slices,labels=activities,colors=['c','m','r','b'])
```

```python
# This pie chart looks bit ugly
# So we can add more features :
plt.pie(slices,labels=activities,colors=['c','m','r','b'],shadow=True , explode = (0.2,0,0,0),autopct='%1.1f%%')
# here explode is used to take a slice of any category out ! We can change it from 0.2 to any number ! 
# Shadow ives a shadow to the chart 
# autopct shows the percentage on the Chart
```

# Plotting 2 graphs at the same time :
We use the plt.subplot to plot multiple graths at the same time 

```python
import matplotlib.pyplot as plt
```

```python
# First graph that we need to print is a pie chart
# So we HAve to use subplot to specify the position of the graph 

plt.subplot(211)
# here 2=we have 2 graphs , 1 = Horizontally only one graph , 1= on vertical position this will be our first graph 
plt.pie(slices,labels=activities,colors=['c','m','r','b'],shadow=True , explode = (0.2,0,0,0),autopct='%1.1f%%')

# Lets also plot another graph 
plt.subplot(212)
# here 2=we have 2 graphs , 1 = Horizontally only one graph , 2= on vertical position this will be our second graph 
plt.bar([1,3,5,7,9],[5,2,7,8,2],label='Example-One')
```

```python
# we can also show them horizontally :

plt.subplot(221)
plt.pie(slices,labels=activities,colors=['c','m','r','b'],shadow=True , explode = (0.2,0,0,0),autopct='%1.1f%%')

# Lets also plot another graph 
plt.subplot(222)
plt.bar([1,3,5,7,9],[5,2,7,8,2],label='Example-One')
```
